```python
import numpy as np
import pandas as pd
```

```python
market=pd.read_csv(r'28-04-2015-TO-29-04-2020BHARTIARTLALLN.csv')
```

```python
market.head()
```

```python
market[["dd", "mm", "yyyy"]] = market["Date"].str.split("-", expand=True)
market['dd']=market['dd'].astype('int64')
market['mm']=market['mm'].astype('int64')
market['yyyy']=market['yyyy'].astype('int64')
market=market.drop(['Symbol','Series','Date'], axis=1)
market.describe()
```

```python
market["day"] = range(1,len(market.dd)+1)
#from sklearn.pipeline import Pipeline
#from sklearn.impute import SimpleImputer
#from sklearn.preprocessing import StandardScaler
#pl=Pipeline([('imp',SimpleImputer(strategy='median')),('scale',StandardScaler())])
#m=pl.fit_transform(market)
#market=pd.DataFrame(m,columns=market.columns)
market.corr()
```

```python
x=market[['Prev Close']]
y1=market[['Open Price']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,y1train,y1test=train_test_split(x,y1,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
#omodel=RandomForestRegressor()
#omodel=DecisionTreeRegressor()
omodel=LinearRegression()
omodel.fit(xtrain,y1train)
```

```python
y1predict=omodel.predict(xtest)
```

```python
xtest
print(xtest[10:20],
    list(y1predict[10:20]),
      (y1test[10:20]))
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y1test,y1predict)
rmse=np.sqrt(mse)
rmse
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores=cross_val_score(omodel,y1test,y1predict,scoring='neg_mean_squared_error')
rmse_scores=np.sqrt(-scores)
```
